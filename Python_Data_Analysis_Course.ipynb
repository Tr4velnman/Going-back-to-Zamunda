```python
#Import Numpy
import numpy as np
```

```python
#Create numpy array
np1 = np.array([0,1,"bob",3,4,5,6,7,8,9])
print(np1)
print(np1[2])
```

```python
# Shape - shows the number of elements in the array like the python len function
print(np1.shape)
```

```python
# Arange
np2 = np.arange(10)
print(np2)
```

```python
# Step
np3 = np.arange(0, 100, 2)
print(np3)
```

```python
# Zeros
np4 = np.zeros(10)
print(np4)
```

```python
# Multi-Dimensional Arrays with Zeros
np5 = np.zeros((5, 10))
print(np5)
```

```python
# Full
np6 = np.full((10), "Bob")
print(np6)
```

```python
# Multi-Dimensional Arrays with Full
np7 = np.full((2, 10), "Bob")
print(np7)
```

```python
# Convert Python List to Numpy Array
# my_list
my_list = ["Bob", "Justin"]
np8 = np.array(my_list)
print(np8)
print(my_list)
```
